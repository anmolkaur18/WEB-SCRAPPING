```python
soup = BeautifulSoup(data.content,'html.parser')
```

```python
soup
```

```python
soup.find('title').text.strip()
```

```python
abc = soup.find('div',attrs ={'id':'more-news'})
```

```python
abc1 = abc.find_all('li')
abc2 = abc.ul.find_all('span')
gh = []
for ab,cd in zip(abc1,abc2):
    ab = ab.text.strip().replace(cd.text,'').strip()
    gh.append(ab)
    # print ab

for i in range(0,10):
    print(gh[i])
```

```python
def remove_space(te):
    return ' '.join(te.split())
```
